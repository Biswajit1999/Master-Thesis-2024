```python
import serial
import matplotlib.pyplot as plt

# Configure the serial port
ser = serial.Serial('COM4', 9600, timeout=1)

# Define the command codes
N = b'GN00001'
S = b'GS00001'
E = b'GT00001'
W = b'GW00001'
HANDSHAKE = b'V'
RETURN = b'R'
stepNorth = b'N01111'  # 15 steps north
stepEast = b'E00101'    # 5 steps east
stepSouth = b'S01111'  # 15 steps South

# Initialize the position
x, y = 0, 0
positions = [(x, y)]  # List to store positions

# Create a figure and axis for plotting
fig, ax = plt.subplots(figsize=(8, 6))
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title('SX-AO-USB Unit Movement')
ax.axhline(y=0, color='k', linestyle='--')
ax.axvline(x=0, color='k', linestyle='--')
ax.set_xlim([-10, 10])
ax.set_ylim([-10, 10])

# Label the directions
ax.text(8, 0, 'East', ha='left', va='top', fontsize=14, color='r')
ax.text(-8, 0, 'West', ha='right', va='top', fontsize=14, color='g')
ax.text(0, 8, 'North', ha='center', va='bottom', fontsize=14, color='b')
ax.text(0, -8, 'South', ha='center', va='top', fontsize=14, color='m')

# Define the movement operations
operations = ['K', 'E', 'W', 'W', 'S', 'S', 'E', 'E', 'E', 'N', 'N', 'N', 'W', 'W', 'W', 'W', 'S', 'S', 'S', 'S', 'E', 'E', 'E', 'E', 'N', 'N','R']

for op in operations:
    if op == 'K':
        print("Finding center")
        ser.write(b'K')
        response = ser.readline().decode().strip()
        print(f"Response: {response}")
    elif op == 'N':
        print("Moving north (up) one step")
        ser.write(N)
        response = ser.readline().decode().strip()
        print(f"Response: {response}")
        y += 1
    elif op == 'S':
        print("Moving south one step")
        ser.write(S)
        response = ser.readline().decode().strip()
        print(f"Response: {response}")
        y -= 1
    elif op == 'E':
        print("Moving east one step")
        ser.write(E)
        response = ser.readline().decode().strip()
        print(f"Response: {response}")
        x += 1
    elif op == 'W':
        print("Moving west one step")
        ser.write(W)
        response = ser.readline().decode().strip()
        print(f"Response: {response}")
        x -= 1
    elif op == 'R':
        print("Centering at low speed")
        ser.write(RETURN)
        response = ser.readline().decode().strip()
        print(f"Response: {response}")
    elif op == 'stepNorth':
        print("Moving 15 steps north")
        ser.write(stepNorth)
        response = ser.readline().decode().strip()
        print(f"Response: {response}")
        y += 1
    elif op == 'stepEast':
        print("Moving 5 steps east")
        ser.write(stepEast)
        response = ser.readline().decode().strip()
        print(f"Response: {response}")
        x += 1
    elif op == 'stepSouth':
        print("Moving 15 steps south")
        ser.write(stepSouth)
        response = ser.readline().decode().strip()
        print(f"Response: {response}")
        y -= 1
    else:
        print("Invalid operation")

    # Update the position list
    positions.append((x, y))

    # Plot the current position
    if len(positions) == 1:
        ax.plot(x, y, 'ro', markersize=6)  # Plot the initial point in red
    else:
        ax.plot([positions[-2][0], x], [positions[-2][1], y], '-.')  # Plot the path

ser.close()
plt.show()
```
